```python
model = "qwen2.5:iq4"
```

```python
import requests
prompt = "Why is the sky blue? Wrong answers only."
resp = requests.post('http://localhost:11434/api/generate', json={
    "model": model,
    "prompt": prompt,
    "stream": False,
})
text = resp.json()["response"]
print(text)
```

```python
import ollama

stream = ollama.chat(
    model=model,
    messages=[{'role': 'user', 'content': 'Why is the sky blue?'}],
    stream=True,
)

for chunk in stream:
  print(chunk['message']['content'], end='', flush=True)
```
